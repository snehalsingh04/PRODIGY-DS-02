# Perform data cleaning and exploratory data analysis (EDA) on a dataset of your choice, such as the Titanic dataset from Kaggle. Explore the relationships between variables and identify patterns and trends in the data.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df
```

### Performing Data Cleaning such as handling missing data

```python
#Finding missing data
df.isnull().sum()
```

```python
#Data Cleaning on Age,Cabin,Embarked
#filling missing data by the median of ages
df['Age'].fillna(df['Age'].median(),inplace=True)

#filling missing data by the mode of embarked
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)

#droping the cabin column due to large number of missing data
df.drop(columns='Cabin',inplace=True)
```

```python
df.isna().sum()
```

### Explore the relationships between variables and identify patterns and trends in the data.


```python
#Using Seaborn library and Matplotlib Library for this
#1.Survival Rate by Passenger Class on the basis of Sex

sns.barplot(x='Pclass', y='Survived', data=df,hue='Sex')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#2.Gender by Age on the basis of Passenger Class

sns.barplot(x='Sex', y='Age', data=df,hue='Pclass')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
#3. Gender by frequency on the basis of Passenger Class

sns.countplot(data=df,x='Sex',hue='Pclass')
plt.title('Frequency of Gender in Passenger Class')
plt.show()
```

```python
#4.Passenger Class by Fare on the basis of Gender

sns.barplot(x='Pclass', y='Fare', data=df,hue='Sex')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```
